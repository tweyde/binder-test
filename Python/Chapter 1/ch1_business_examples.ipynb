```python
# Data structure
import pandas as pd

# Charts...
import matplotlib.pyplot as plt
```

## Use case: Correlations to Determine US Supermarket Sales Strategy

```python
#Get the data ... 
df_supermarket = pd.read_csv("..\\Data\\50_SupermarketBranches.csv")
df_supermarket.head()
```

```python
# Extract columns to do the analysis
x_1 = df_supermarket['Promotion Spend']
x_2 = df_supermarket['Advertisement Spend']
y = df_supermarket['Profit']
```

```python
# use Scatter to show the relationship between profit and promotion spend by product line
plt.scatter(x_1,y)
plt.title('Relationship between Promotion Spend and profit (each point is a product line)');
plt.xlabel('Promotion Spend')
plt.ylabel('Profit')
plt.show()
```

```python
# use Scatter to show the relationship between profit and promotion spend by product line
plt.scatter(x_2,y)
plt.title('Relationship between Advertisement Spend and profit (each point is a product line)');
plt.xlabel('Advertisement Spend')
plt.ylabel('Profit')
plt.show()
```

```python
# Examine correlations.
# This function calculates the correlation of each column against each other
# Look up the correlation between two columns to see how good it is. For example, 
# note that the correlation between Profit and Advertisement Spend is high (0.97) 
# , where as the correlation between Profit and Promotion Spend is low (0.20)
corr = df_supermarket.corr(method ='pearson')
corr
```

```python
# Sort in order of correlation with Profit.... (clearly profit will have the highest)
corr.sort_values(['Profit'], ascending=False)[['Profit']]
```
