```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Exercises

1. Calculate the average life expectancy for each year.

```python
gapminder.groupby(['year'])['lifeExp'].mean()
```

2. Plot year versus life expectancy and draw a horizontal line at the average life expectancy value. Color the points based on the continent

```python
plt.style.use('default')
sns.scatterplot(x='year', y='lifeExp', hue='continent', data=gapminder)
plt.axhline(gapminder['lifeExp'].mean(), color='blue')
plt.show()
```

3. Calculate the total population for each continent for each year. Create an appropriate plot to visualize the data

```python
gapminder.groupby(['year'])['pop'].sum()
```

```python
plt.bar(gapminder['year'], gapminder['pop'])
plt.show()
```

4. Plot GDP per capita for each year for African countries. Add appropriate labels to indicate the country name.

```python
g = gapminder[gapminder['continent']=='Africa']
g.head()
```

```python
sns.lineplot(x='year', y='gdpPercap', hue='country', data=g, palette='deep')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()
```
